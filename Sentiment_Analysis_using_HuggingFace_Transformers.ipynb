```python
from transformers import pipeline

# Load the pre-trained sentiment-analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Input sentence
sentence = "Despite the high price, the performance of the new MacBook is outstanding."

# Perform sentiment analysis
result = sentiment_analyzer(sentence)[0]

# Print the label and confidence score
print(f"Sentiment: {result['label']}")
print(f"Confidence Score: {result['score']:.4f}")
```

1. What is the main architectural difference between BERT and GPT? Which uses an encoder and which uses a decoder?
BERT (Bidirectional Encoder Representations from Transformers) uses only encoders and is trained in a bidirectional manner, meaning it considers context from both sides of a word.

GPT (Generative Pre-trained Transformer) uses only decoders and is trained in a unidirectional (left-to-right) manner.

2. Why is using pre-trained models (like BERT or GPT) beneficial for NLP applications instead of training from scratch?
Pre-trained models provide a strong starting point by leveraging knowledge from large datasets. They save time, resources, and computational power, while also delivering better performance on downstream tasks by fine-tuning on smaller task-specific datasets.
